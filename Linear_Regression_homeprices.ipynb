```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/folder/homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[1500]])
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
